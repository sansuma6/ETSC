````python
import pandas as pd
url="https://raw.githubusercontent.com/GLambard/Molecules_Dataset_Collection/master/latest/ESOL_delaney-processed.csv"
df = pd.read_csv(url, index_col=0)
df
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pairwise Correlation Heatmap')
plt.show()
````

````python
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(df['Molecular Weight'], bins=20, color='blue', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Molecular Weight')
plt.ylabel('Frequency')
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the columns you want to create histograms for
selected_columns = ['Molecular Weight', 'Number of H-Bond Donors', 'Number of Rings']

# Create histograms
for column in selected_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=column, kde=True, bins=20)  # Adjust the number of bins as needed
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
````
